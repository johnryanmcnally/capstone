```python
import altair as alt
import pandas as pd
```

```python
modelpath = 'models/'
n = 5
df = pd.read_csv(modelpath+'tsne_plot/tsnedf_{}.csv'.format(15))
labelout = ''
count = 0
for idx, row in df.iterrows():
    labelin = row.label
    if labelin != labelout:
        count = 0
    count+=1
    if count >= 5:
        df.at[idx,'word'] = ''
    labelout = row.label
df.head(10)
```

```python
title = alt.TitleParams('2D t-SNE of Ethnicity Topic Vectors',
 subtitle='Text Labels Added to Top 5 Tokens',
 anchor='middle',
 fontSize=20,
 fontWeight='bold'
)
c = alt.Chart(df, title=title).mark_circle(size=60,stroke='black', strokeWidth=.25).encode(
    x = alt.X('x:Q', axis=None),
    y = alt.Y('y:Q', axis=None),
    color = 'label:N',
    #tooltip = ['word','label',],
    )        
t = c.mark_text(dy=8, fontSize=12).encode(
    text='word:N',
    )

(c+t).configure_axis(
    grid=False
    ).configure_legend(
        labelFontSize=14,
        orient='right',
    ).configure_view(
        strokeWidth = 0
    ).properties(
        height=600,
        width = 800
    ).interactive()
```
